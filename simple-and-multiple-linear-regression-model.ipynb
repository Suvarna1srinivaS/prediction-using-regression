<font color = "#FF5F1F">1. Linear Regression Model:
--------------------

This is a supevised learning approach in which  dependent vraible should be continuous. It shows the relationship between independent and dependent variable.






<font color = "#FF5F1F">a.   Simple linear regression:  </font>


This is a univariate linear regresion that assess the relation between dependent and independent variable.   
One independent variable and dependent variable are considered to analyse the dataset.  
The simple linear model is expressed using the following equation:  
    y = a + bx + e, where

   *   a = intercept (distance of best fit line from x-axis where it touches the x-axis)
   *   b = slope (angle made by the line)
   *   e = residual or error (distance between the datapoint and the predicted line)

<font color = "#FF5F1F">Ordinary Least Square technique (OLS):

This technique is used in linear regresson models to estimate the coefficients of linear equation.   
This technique tries to minimize the sum of squared residuals between the actaul and predicted values.
It will draw multiple lines and shows the line which covers the maximum data points (distance from the data to the line should be less)


*   Advantages of Ordinary Least Square technique:

 1.   It works with numeric attributes
 2.   Model learning time is fast
 3.   Prediction is powerful if the dimension is high

*   Disadvantages of Ordinary Least Square technique:

 1.   For high dimensional data, model prone to overfitting
 2.   Cannot perform variable selection
 3.   Sensitive to outliers






<font color = "#FF5F1F">Assumptions for Linear Regression Model:


*   Before model building:

 1.   Y must be normally distributed
 2.   There must be a linear relationship between x and y


*   After model building:

 3.   There should not be multicollinearity in the data
 4.   There should not be heteroscedasticity
 5.   There should not be auto correlation in the data
 6.   Residuals should also be normally distributed



<font color = "#FF5F1F">Evaluation:


Once the optimized algorithm builds the model, it is crucial to assess the quality or performance of its statistical modeling.  
The evaluation of the model must be done on a new data set with labels that were not part of the optimization process

Measures:

The central quality measure for regression is root mean square erroe (RMSE)  
It measures the residuals (e) and penalizes large errors stronger than good predictions.  


Error measures can be measured and accumulated in various ways:
*   Absolute score:
  1.   Mean Square error (quadratic)
  2.   Root Mean Square error (quadratic)
  3.   Mean Absolute error (linear)

*   Relative score error:
  1. Mean Absolute Percentage error (linear)
  2. R-square (co-efficient of determination)



where,  
linear - equal weighting to errors  
quadratic - stronger penalty for larger errors

Note: All measures have a range (0, infinity)

R-square:


                           sum of squares of residuals
            R-square = 1 - ---------------------------
                              total sum of squares


  if R-square < 0, model performs bad  
  if R-square = 1, model performs perfectly

  *   Range of R-square (-infinity, 1)
  *   for a real world noisy datasets, R-square values above 0.4 are often considered okay
  *   for a good model, R-square and adjacent R-square should be very close.  


The major difference between R-square and adj R-square is "The R-square of a model increases if we add another independent variable to the model. But the adj R-square value increases only if the added independent variable has any effect on dependent variable".

```python
import pandas as pd
car_df = pd.read_csv('/kaggle/input/regression-data/regression data.csv')
```

```python
car_df = car_df.drop(["Unnamed: 0"], axis = 1)
```

```python
car_df.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(car_df)
scaled_df = pd.DataFrame(scaled_data, columns = car_df.columns)
scaled_df.sample(2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (15,7))
sns.heatmap(scaled_df.corr(), annot = True, cmap = "viridis")
```

```python
main_df = scaled_df[['cyl', 'disp', 'hp', 'wt', 'carb', 'mpg']]
```

```python
import numpy as np
from sklearn.model_selection import train_test_split


y = np.array(car_df["mpg"])
x = np.array(main_df["wt"]).reshape((-1,1))


x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.3, random_state = 200)
```

```python
from sklearn.linear_model  import LinearRegression

model = LinearRegression()

z = model.fit(x_train,y_train) 
intercept = z.intercept_
coefficient = z.coef_
print("intercept: ",intercept)
print("coefficient: ",coefficient)
```

### <font color = "0096FF">R-Square value for train data:

```python
r_sqr_i_train = z.score(x_train, y_train)
print(r_sqr_i_train)
```

### <font color = "0096FF">R-Square value for test data:

```python
r_sqr_i_test = z.score(x_test, y_test)
print(r_sqr_i_test)
```

### <font color = "0096FF"> predicting values & calculating errors using train data:

```python
predicted_train_y = z.predict(x_train)

from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

MSE = mean_squared_error(y_train, predicted_train_y)
MAE = mean_absolute_error(y_train, predicted_train_y)
RMSE = mean_squared_error(y_train, predicted_train_y, squared = False)

print("mean square error: ", MSE)
print("mean absolute error: ", MAE)
print("root mean square error: ", RMSE)
```

### <font color = "0096FF"> predicting values & calculating errors using test data:

```python
predicted_test_y = z.predict(x_test)

MSE = mean_squared_error(y_test, predicted_test_y)
MAE = mean_absolute_error(y_test, predicted_test_y)
RMSE = mean_squared_error(y_test, predicted_test_y, squared = False)

print("mean square error: ", MSE)
print("mean absolute error: ", MAE)
print("root mean square error: ", RMSE)
```

<font color = "#FF5F1F">b.   Multiple Linear Regression:


Multiple linear regression is similar to the simple linear regression with the exception that multiple independent variables are used in the model to assess one dependent variable.  
The Multiple linear model is expressed using the following equation:  
    y = a + b1x1 + b2x2 + b3x3 + e, where

   *   a = intercept (distance of best fit line from x-axis where it touches the x-axis)
   *   b1, b2, b3 = slopes (angles made by the line)
   *   e = residual or error (distance between the datapoint and the predicted line)

<font color = "#FF5F1F"> Assumptions of Multiple Linear Regression:</font>  
Assumptions are same as Linear Regression Model, But along with them:
1.   Non-collinearity
2.   Residuals must be normally distributed

<font color = "#FF5F1F"> Overfitting and Underfitting:

* if the train error is high and test error is low, model is underfitted 
* if the train error is low and test error is high, model is overfitted (model not able to generalise the data)
* if the train error and test error are close to each other, model is optimal
* if the model has high variance, model is overfitted
* if the model has high bias, model is underfitted
* if the bias and variance are low, model is optimal. This is also called as bias-variance trade-off.

**From the RMSE values of train and test data, I conclude that the test error is high when compared to train error for all the three models I built. Hence, the model is overfitted.**
